```python
# Python program to generate a data set of numbers 1-N (size) with their factors
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

size = 256;
numbers = {};

# creates a list for each number to add factors to, i.e. {1:[],2:[],...}
for n in range(1,size+1):
    numbers[n]=[];

# for each number, add its multiples up to N to the corresponding arrays
# e.g. add 3 to 3,6,9,12,...,255 for N = 256

for j in range(1,size+1):
    for i in range(1,1+size/j):
        numbers[i*j].append(j)

# for each number, get the number of factors
# this can be used to plot the number of factors for each number, and
# to determine if the number is prime (i.e. numfactors = 2)

numfactors = [];
for i in range(1,size+1):
    numfactors.append([i, len(numbers[i]) ]);



#colors = list("rgbcmyk")
colors = list("rc")
#vedgelist = []

for i in range(1,size+1):
    for j in range(1,len(numbers[i])):
        x = i
        y = numbers[i][j]
        # print even factors in red
        color = "r"
        # print odd factors in blue
        if (y%2 ==1):
            color = "c"
            # print prime numbers in yellow
            if (numfactors[j][1]==2):
                color = "y"
        plt.scatter(x,y,color=color)
        
        #store data in different format to make graph (V,E)
        #convert to pairs for vertex and edges
        #if(x!=y):
        #    vedgelist.append([x,y])

#np.savetxt("factors.csv", vedgelist, delimiter=",")	
#np.savetxt("numfactors.csv", numfactors, delimiter=",")

plt.axis([0,size+1,0,size+1])
ax = plt.gca()

ax.set_axis_bgcolor("k")

plt.title("factors")
plt.show()
```
